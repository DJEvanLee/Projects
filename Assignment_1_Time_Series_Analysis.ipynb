```python

df = pd.read_excel(xls, sheet_name=0, skiprows=4)  # Skip first 4 rows as they are not part of the data

# Rename the columns to give them meaningful names
df.columns = ['Category', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df = df.dropna(subset=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], how='all')

# Drop any rows that are not relevant (like separators or sub-headers)
df = df.dropna(how='all')  # Drop rows where all values are NaN



# Display the cleaned dataset
df.head()
```

```python
# Filter the data to get the 'Total costs listed' or other specific rows you are interested in
total_costs_df = df[df['Category'].str.contains('Total costs listed', na=False)]

# Reset the index for better readability
total_costs_df.reset_index(drop=True, inplace=True)

# Convert the month columns to numeric values
total_costs_df.iloc[:, 1:] = total_costs_df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# Display the cleaned and filtered total costs data
total_costs_df
```

```python
# Melt the dataframe to create a time series format (Month-Year format)
total_costs_melted = total_costs_df.melt(id_vars=['Category'], var_name='Month', value_name='Cost')

# Add a year column (assuming you only have data for 2005 here)
total_costs_melted['Year'] = 2005

# Combine Year and Month to create a datetime index
total_costs_melted['Date'] = pd.to_datetime(total_costs_melted['Year'].astype(str) + '-' + total_costs_melted['Month'])

# Set the Date as the index and sort the data
total_costs_melted.set_index('Date', inplace=True)
total_costs_melted = total_costs_melted.sort_index()

# Drop unnecessary columns like 'Category' and 'Year'
total_costs_melted.drop(['Category', 'Year'], axis=1, inplace=True)

# Display the cleaned, time-series formatted dataset
total_costs_melted.head()
```

```python
# Plot the time series of the total costs
plt.figure(figsize=(10, 6))
plt.plot(total_costs_melted.index, total_costs_melted['Cost'], label='Total Costs Listed')
plt.title('Monthly Dairy Costs of Production per cwt of Milk Sold (2005)')
plt.xlabel('Date')
plt.ylabel('Cost ($/cwt)')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Apply Moving Average with different window sizes
total_costs_melted['MA_3'] = total_costs_melted['Cost'].rolling(window=3).mean()
total_costs_melted['MA_6'] = total_costs_melted['Cost'].rolling(window=6).mean()
total_costs_melted['MA_12'] = total_costs_melted['Cost'].rolling(window=12).mean()

# Plot the original series and the smoothed series for each window size
plt.figure(figsize=(12, 8))
plt.plot(total_costs_melted.index, total_costs_melted['Cost'], label='Original', color='blue', alpha=0.5)
plt.plot(total_costs_melted.index, total_costs_melted['MA_3'], label='Moving Average (3 months)', color='red')
plt.plot(total_costs_melted.index, total_costs_melted['MA_6'], label='Moving Average (6 months)', color='green')
plt.plot(total_costs_melted.index, total_costs_melted['MA_12'], label='Moving Average (12 months)', color='orange')

plt.title('Moving Average with Window Sizes 3, 6, 12')
plt.xlabel('Date')
plt.ylabel('Cost ($/cwt)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing
# Ensure that the 'Cost' column is numeric
total_costs_melted['Cost'] = pd.to_numeric(total_costs_melted['Cost'], errors='coerce')

# Check for any missing values and fill or drop them as needed
total_costs_melted['Cost'].fillna(method='ffill', inplace=True)  # Forward fill any missing values

# If forward fill isn't appropriate, you can also use backward fill
# total_costs_melted['Cost'].fillna(method='bfill', inplace=True)
# Apply Single Exponential Smoothing
model_single = SimpleExpSmoothing(total_costs_melted['Cost']).fit(smoothing_level=0.2, optimized=False)
total_costs_melted['SES'] = model_single.fittedvalues

# Apply Double Exponential Smoothing (Holt's Linear)
model_double = ExponentialSmoothing(total_costs_melted['Cost'], trend='add').fit()
total_costs_melted['DES'] = model_double.fittedvalues

# Plot the original series and the forecasts
plt.figure(figsize=(12, 8))
plt.plot(total_costs_melted.index, total_costs_melted['Cost'], label='Original', color='blue', alpha=0.5)
plt.plot(total_costs_melted.index, total_costs_melted['SES'], label='Single Exponential Smoothing', color='red')
plt.plot(total_costs_melted.index, total_costs_melted['DES'], label='Double Exponential Smoothing', color='green')

plt.title('Single and Double Exponential Smoothing')
plt.xlabel('Date')
plt.ylabel('Cost ($/cwt)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Drop NaN values and align the data by trimming to the same length
cost_ma3 = total_costs_melted[['Cost', 'MA_3']].dropna()
cost_ma6 = total_costs_melted[['Cost', 'MA_6']].dropna()
cost_ma12 = total_costs_melted[['Cost', 'MA_12']].dropna()

# Calculate MAE and MSE for Moving Averages with aligned data
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Moving Average with 3-month window
mae_ma3 = mean_absolute_error(cost_ma3['Cost'], cost_ma3['MA_3'])
mse_ma3 = mean_squared_error(cost_ma3['Cost'], cost_ma3['MA_3'])

# Moving Average with 6-month window
mae_ma6 = mean_absolute_error(cost_ma6['Cost'], cost_ma6['MA_6'])
mse_ma6 = mean_squared_error(cost_ma6['Cost'], cost_ma6['MA_6'])

# Moving Average with 12-month window
mae_ma12 = mean_absolute_error(cost_ma12['Cost'], cost_ma12['MA_12'])
mse_ma12 = mean_squared_error(cost_ma12['Cost'], cost_ma12['MA_12'])

# Print the results
print(f'MA (3 months): MAE = {mae_ma3:.4f}, MSE = {mse_ma3:.4f}')
print(f'MA (6 months): MAE = {mae_ma6:.4f}, MSE = {mse_ma6:.4f}')
print(f'MA (12 months): MAE = {mae_ma12:.4f}, MSE = {mse_ma12:.4f}')
```

Summary:

Moving Average Forecasting was applied with window sizes of 3, 6, and 12 months. The 3-month moving average was the most effective, with the lowest Mean Absolute Error (MAE = 0.3418) and Mean Squared Error (MSE = 0.1607), capturing short-term trends well. The 12-month moving average smoothed out long-term trends but lagged behind recent changes.

Exponential Smoothing methods were also used. Double Exponential Smoothing (DES) performed better than Single Exponential Smoothing (SES) due to its ability to model both level and trend, making it more responsive to recent changes.

Conclusion: The 3-month moving average is best for short-term forecasting due to its accuracy, while Double Exponential Smoothing effectively captures long-term trends. Together, these techniques provide a comprehensive approach for analyzing the dataset.
